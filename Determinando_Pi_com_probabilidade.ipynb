```python
import numpy as np # Trabalhar com arrays
import matplotlib.pyplot as plt # Plotagem de gráficos
import math as math # Funções matemáticas não nativas
```

```python
# Parametros iniciais para vetor de números aleatórios
intervalo = 5
qnt_inicial = 5
qnt_final = 1000
amostras = (qnt_final - qnt_inicial)//5

# Vetores de números aleatórios
x = np.round(np.random.rand(amostras),4)
y = np.round(np.random.rand(amostras),4)
```

```python
# Informações sobre a circunferencia
raio = 0.5

a = 0.5 # Coordenada X do Centro
b = 0.5 # Coordenada Y do Centro

# Variáveis que receberão pontos dentro e fora da circunferencia
dentro = 0
fora = 0
```

```python
# Verificação de pontos dentro e fora da circunferencia
for i in range(amostras):
    ddp = math.sqrt((x[i] - a)**2 + (y[i] - b)**2)
    
    if ddp > raio:
        fora += 1
    else:
        dentro += 1 
```

```python
# Plotagem dos pontos junto a circunferencia

plt.figure(figsize=(6, 6))

plt.plot(x, y, 'ro', label='Pontos')

circle = plt.Circle((a, b), raio, edgecolor='blue', facecolor='none')
plt.gca().add_patch(circle)

plt.xlim(0, 1)
plt.ylim(0, 1)

plt.show()
```
